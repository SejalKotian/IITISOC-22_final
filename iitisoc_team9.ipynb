Imported all the necessary libaries and defined the path

```python
def image_feature(path):
    model = tf.keras.applications.InceptionResNetV2(weights='imagenet',include_top= False)
    features=[];
    img_name=[];
    for folder in os.listdir(path):
        sub_path=path+"/"+folder
        for img in os.listdir(sub_path):
            image_path=sub_path+"/"+img
            img_arr=cv2.imread(image_path)
            img_arr=cv2.resize(img_arr,(150,150))
            img_arr=np.expand_dims(img_arr,axis=0)
            img_arr=preprocess_input(img_arr)
            feat=model.predict(img_arr)
            feat=feat.flatten()
            features.append(feat)
            img_name.append(img)
    return features,img_name
```

* Extracted the feature of the image using pre-trained model InceptionResNetV2
* Feature Extraction:
* 1,Instantiated a base model and loaded the pre-trained weights into it.
* 2.Ran our dataset through it and recorded the output  of the layers (excluding the top layer) to extract the features of each image.

```python
path= '../input/datasetiitisocac'
img_features,img_names=image_feature(path)
```

```python
import pandas as pd
```

```python
make_list=pd.DataFrame(list(zip(img_names,img_features)),columns=['Img_Names','Img_features'])
```

**Created a dataframe of two columns with img names against image features **

```python
shuffled= make_list.sample(frac=1)
print(shuffled.head(20))
```

```python
shuffled_list= list(shuffled['Img_features'])
shuffled_list[0]
```

Shuffled the dataframe so that the model does not learn anything from the order of the images passed


```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(shuffled_list)
shuffled_list_sc = sc.transform(shuffled_list)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
shuffled_list_pca=pca.fit_transform(shuffled_list_sc)
```

Applied PCA(Principal Component Analysis) to improvise the model

```python
#Creating Clusters
#Implementing kMeans clustering algo
k = 10
clusters= KMeans(n_clusters=10, random_state=0, init='k-means++', max_iter=300, n_init=10, tol=0.0001,copy_x=True,  algorithm ='auto')

labels=clusters.fit_predict(shuffled_list_pca)
labels
```

```python
img_names_shuffled_list= list(shuffled['Img_Names'])
image_cluster = pd.DataFrame(img_names_shuffled_list,columns=['image'])
image_cluster["clusterid"] = clusters.labels_
image_cluster 
```

```python
from sklearn.metrics import silhouette_score
```

```python
score = silhouette_score(shuffled_list_pca, labels, metric='euclidean')
```

```python
score
```

A negative silhouette score indicates wrong clustering 
Whilst,a spositive silhouette score indicates correct clustering
